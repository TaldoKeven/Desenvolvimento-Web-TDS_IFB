### O que é um Sistema TODO?

Um sistema "TODO" (do inglês "to do", que significa "para fazer") é essencialmente uma lista de tarefas. É uma ferramenta projetada para ajudar a organizar e acompanhar as atividades que precisam ser concluídas.

No contexto de sistemas web e desenvolvimento, um sistema TODO geralmente implementa as operações básicas de gerenciamento de dados:

*   **Criação (Create):** Adicionar novas tarefas à lista.
*   **Leitura/Visualização (Read):** Exibir as tarefas existentes.
*   **Atualização (Update):** Modificar o status (como marcar como concluída), prioridade ou outros detalhes de uma tarefa.
*   **Remoção (Delete):** Excluir tarefas que não são mais relevantes.

Neste notebook, simulamos um sistema TODO simples utilizando uma lista de dicionários em Python, onde cada dicionário representa uma tarefa individual.

```
# Nossa lista principal que armazena todas as tarefas.
# Cada elemento da lista é um dicionário.
lista_de_tarefas = []

print("--- 1. Inicialização do Gerenciador de Tarefas ---")
```

O uso de **listas de dicionários** nesta atividade é a simulação  de como os dados são organizados no *backend* de sistemas web.

## Roteiro da Aula: Gerenciador de Tarefas (TODO List)

### 1\. Preparação: Inicialização (Revisão da Estrutura)

**Conceitos Revisados:**

Criação de lista (`[]`) e inserção de elementos (dicionários).

**Passo 1.1: Criar a Lista Principal**

Definimos a lista que funcionará como nosso "Banco de Dados" temporário.

**Passo 1.2: Adicionar Tarefas (CRUD: Create)**

Adicionamos as tarefas, simulando a criação de novos registros no sistema.

```
# Adicionando tarefas iniciais (Simulando a criação de dados)
lista_de_tarefas.append({
    'id': 1,
    'descricao': 'Implementar o endpoint de login (Exemplo 2: Rotas)',
    'prioridade': 'Alta',
    'concluida': False
})

lista_de_tarefas.append({
    'id': 2,
    'descricao': 'Documentar a API de produtos (Exemplo 1: Carrinho)',
    'prioridade': 'Baixa',
    'concluida': False
})

lista_de_tarefas.append({
    'id': 3,
    'descricao': 'Testar filtro de usuários ativos (Exemplo 3: Painel)',
    'prioridade': 'Alta',
    'concluida': False
})

print(f"Total de tarefas criadas: {len(lista_de_tarefas)}")
```

### 2\. Visualização e Filtragem (Revisão de Loops e Filtros)

**Conceitos Revisados:**

Iteração com `for`, acesso a valores em dicionários, e **filtragem** condicional.


**Passo 2.1: Exibir Todas as Tarefas**

Função auxiliar para exibir a lista formatada.

```
def exibir_tarefas(titulo, lista):
    print(f"\n--- {titulo} ({len(lista)} tarefas) ---")
    for t in lista:
        status = "✅ CONCLUÍDA" if t['concluida'] else "❌ PENDENTE"
        print(f"ID: {t['id']} | Prioridade: {t['prioridade']:5s} | Status: {status} | Descrição: {t['descricao']}")

exibir_tarefas("Tarefas Atuais", lista_de_tarefas)
```

**Passo 2.2: Filtrar Pendentes (Foco no Exemplo 3)**

Isolamos apenas as tarefas que ainda não foram concluídas.

```
# Filtrando dados (Revisão do Exemplo 3: Filtragem)
# Utilizamos list comprehension (ou um loop if/else) para criar uma nova lista.
tarefas_pendentes = [
    tarefa for tarefa in lista_de_tarefas if tarefa['concluida'] == False
]

exibir_tarefas("Tarefas Pendentes", tarefas_pendentes)
```

### 3\. Atualização de Status (CRUD: Update e Busca)

**Conceitos Revisados:**

Busca de elemento na lista e modificação (Update) de valores de dicionário.

**Passo 3.1: Marcar Tarefa como Concluída**

Simulamos a alteração do status de uma tarefa pelo seu ID.

```
id_para_concluir = 1
print(f"\n--- 3. Atualizando a Tarefa ID {id_para_concluir} ---")

# Iteramos na lista para encontrar o item correto.
for tarefa in lista_de_tarefas:
    # Busca pelo ID (Revisão do Exemplo 2: Busca por Rota)
    if tarefa['id'] == id_para_concluir:
        tarefa['concluida'] = True # O Update ocorre aqui!
        print(f"Status da tarefa '{tarefa['descricao']}' alterado para CONCLUÍDA.")
        break # É importante sair do loop assim que a tarefa for encontrada.

exibir_tarefas("Tarefas Após a Conclusão", lista_de_tarefas)
```

### 4\. Remoção e Ordenação (CRUD: Delete e Desafio Extra)

**Conceitos Revisados:**

Remoção de elementos (`.pop()` ou `del`) e uso de `sorted()` com `key=lambda` para ordenação.

**Passo 4.1: Remover uma Tarefa (CRUD: Delete)**

Simulamos a exclusão de uma tarefa finalizada ou cancelada.

```
id_para_remover = 2
print(f"\n--- 4. Removendo a Tarefa ID {id_para_remover} ---")

# Encontrar o índice (posição) da tarefa na lista para usar o método .pop()
indice_para_remover = -1
for i, tarefa in enumerate(lista_de_tarefas):
    if tarefa['id'] == id_para_remover:
        indice_para_remover = i
        break

if indice_para_remover != -1:
    # Remove o item pelo seu índice (posição) na lista.
    tarefa_removida = lista_de_tarefas.pop(indice_para_remover)
    print(f"Tarefa removida: '{tarefa_removida['descricao']}'")
else:
    print(f"Erro: Tarefa com ID {id_para_remover} não encontrada.")

exibir_tarefas("Tarefas Após a Remoção", lista_de_tarefas)
```

**Passo 4.2: Ordenação por Prioridade (Desafio, Foco no Exemplo 3)**

Ordenamos a lista final para exibir primeiro as tarefas de **Prioridade Alta**.

```
# Mapeamento para garantir que 'Alta' venha antes de 'Baixa' na ordenação.
# Python ordena strings alfabeticamente, então criamos um valor numérico para forçar a ordem.
prioridade_ordem = {'Alta': 1, 'Média': 2, 'Baixa': 3}

# Usamos sorted() com lambda para ordenar a lista.
# A função key acessa o dicionário 'prioridade_ordem' para pegar o número de ordenação.
tarefas_ordenadas = sorted(
    lista_de_tarefas,
    key=lambda t: prioridade_ordem[t['prioridade']]
)

exibir_tarefas("Lista Final Ordenada por Prioridade", tarefas_ordenadas)
```

Com este roteiro, aplicamos **conceitos** de um projeto de lista de tarefas, que é a base de muitos sistemas web.

```
# ==========================================
# GERENCIADOR DE ATIVIDADES DIÁRIAS - PROGRAMADOR
# Versão para Google Colab / Python
# ==========================================

# --- 1. Inicialização ---
print("1. Inicialização do Gerenciador de Atividades Diárias")

atividades = []

# --- 2. Cadastro de Atividades (CREATE) ---
atividades.append({
    'id': 1,
    'descricao': 'Revisar pull requests e corrigir comentários de código',
    'prioridade': 'Alta',
    'concluida': False
})
atividades.append({
    'id': 2,
    'descricao': 'Implementar nova funcionalidade no módulo de autenticação',
    'prioridade': 'Média',
    'concluida': False
})
atividades.append({
    'id': 3,
    'descricao': 'Atualizar documentação técnica no repositório do projeto',
    'prioridade': 'Baixa',
    'concluida': False
})
atividades.append({
    'id': 4,
    'descricao': 'Participar da reunião de alinhamento com o time de frontend',
    'prioridade': 'Média',
    'concluida': False
})

print(f"Total de atividades cadastradas: {len(atividades)}")

# --- 3. Função para exibir atividades (READ) ---
def exibir_atividades(titulo, lista):
    print(f"\n--- {titulo} ({len(lista)} atividades) ---")
    for a in lista:
        status = "Concluída" if a['concluida'] else "Pendente"
        print(f"ID: {a['id']} | Prioridade: {a['prioridade']:5s} | Status: {status}")
        print(f"Descrição: {a['descricao']}")
        print("-" * 60)

# Exibir todas as atividades
exibir_atividades("Atividades Atuais", atividades)

# --- 4. Filtrar pendentes ---
atividades_pendentes = [a for a in atividades if not a['concluida']]
exibir_atividades("Atividades Pendentes", atividades_pendentes)

# --- 5. Atualizar status (UPDATE) ---
id_para_concluir = 2
print(f"\n5. Atualizando status da atividade ID {id_para_concluir}")

for a in atividades:
    if a['id'] == id_para_concluir:
        a['concluida'] = True
        print(f"A atividade '{a['descricao']}' foi marcada como concluída.")
        break

exibir_atividades("Atividades Após Atualização", atividades)

# --- 6. Remover uma atividade (DELETE) ---
id_para_remover = 3
print(f"\n6. Removendo a atividade ID {id_para_remover}")

indice_para_remover = -1
for i, a in enumerate(atividades):
    if a['id'] == id_para_remover:
        indice_para_remover = i
        break

if indice_para_remover != -1:
    removida = atividades.pop(indice_para_remover)
    print(f"A atividade '{removida['descricao']}' foi removida.")
else:
    print("Erro: Atividade não encontrada.")

exibir_atividades("Atividades Restantes", atividades)

# --- 7. Ordenar por prioridade ---
print("\n7. Ordenando por prioridade")

prioridade_ordem = {'Alta': 1, 'Média': 2, 'Baixa': 3}
atividades_ordenadas = sorted(atividades, key=lambda a: prioridade_ordem[a['prioridade']])

exibir_atividades("Atividades Ordenadas por Prioridade", atividades_ordenadas)
```
